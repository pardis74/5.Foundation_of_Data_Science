```python
def calculate_missing_percentage(data_column):
    total_values = data_column.size
    missing_values = data_column.isna().sum()
    missing_percentage = (missing_values / total_values) * 100
    return missing_percentage

def calculate_cardinality(data_column):
    non_null_values = data_column.dropna()
    unique_values = np.unique(non_null_values)
    cardinality = len(unique_values)
    return cardinality

def calculate_mode_attributes(data_column):
    non_null_values = data_column.dropna()
    
    if non_null_values.empty:
        return None, 0, 0
    
    unique_values, counts = np.unique(non_null_values, return_counts=True)
    sorted_counts = counts.copy()
    sorted_counts.sort()
    
    mode_value = unique_values[counts == sorted_counts[-1]][0]
    mode_count = sorted_counts[-1]
    mode_percentage = (mode_count / len(non_null_values)) * 100
    
    return mode_value, mode_count, mode_percentage

def calculate_2mode_attributes(data_column):
    non_null_values = data_column.dropna()
    
    if non_null_values.empty:
        return None, 0, 0
    
    unique_values, counts = np.unique(non_null_values, return_counts=True)
    sorted_counts = counts.copy()
    sorted_counts.sort()
    
    if len(sorted_counts) < 2:
        return None, 0, 0
    
    second_mode_value = unique_values[counts == sorted_counts[-2]][0]
    second_mode_count = sorted_counts[-2]
    second_mode_percentage = (second_mode_count / len(non_null_values)) * 100
    
    return second_mode_value, second_mode_count, second_mode_percentage


Categorical_Values = []

Categorical_Values = [
    
    {
        "Feature": feat,
        'Desc': features_df.loc[feat]['desc'],
        'Count': values.dropna().size,
        '%Missing': calculate_missing_percentage(values),
        'Card.': calculate_cardinality(values),
        'Mode': calculate_mode_attributes(values)[0],
        'Mode Freq.': calculate_mode_attributes(values)[1],
        'Mode Perc.': calculate_mode_attributes(values)[2],
        '2nd Mode': calculate_2mode_attributes(values)[0],
        '2nd Mode Freq.': calculate_2mode_attributes(values)[1],
        '2nd Mode Perc.': calculate_2mode_attributes(values)[2],
        'Note': ''
    }
    for feat, values in df_cat.items()
]

Categorical_Values_df = pd.DataFrame(Categorical_Values)
Categorical_Values_df
```

### Q3 Outlier Identification (25 pts)

#### Q3.1 For each continuous feature, identify the outliers using the IQR method. (15 pts)
For each feature, report the lower and upper bounds and number of instances that are identified as outliers.
Then, display boxplots and discuss if your outliers analysis makes sense. Discuss why you would (or would not) use the IQR method.


```python
import matplotlib.pyplot as plt

def calculate_iqr_bounds(feat_values):
    q1 = feat_values.quantile(0.25)
    q3 = feat_values.quantile(0.75)
    iqr = q3 - q1
    iqr_lower = q1 - 1.5 * iqr
    iqr_upper = q3 + 1.5 * iqr
    return iqr_lower, iqr_upper

def detect_and_plot_outliers(df, col):
    iqr_lower, iqr_upper = calculate_iqr_bounds(df[col])
    outliers = (df[col] < iqr_lower) | (df[col] > iqr_upper)
    
    print(col)
    print("Upper Bound:", iqr_upper)
    print("Lower Bound:", iqr_lower)
    print("Number of outliers:", outliers.sum())
    
    plt.figure()
    df[[col]].boxplot()

# Iterate through the columns in df_cont
for col in df_cont.columns:
    detect_and_plot_outliers(df_cont, col)

plt.show()
```

Normalize the MARSUPWT, AAGE, NOEMP, CAPGAIN, and CAPLOSS features

* Use Z-score normalization for AAGE feature

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler, FunctionTransformer

columns_to_normalize = ['MARSUPWT', 'AAGE', 'NOEMP', 'CAPGAIN', 'CAPLOSS']

#Create a transformer
scalers = {
    'MARSUPWT': MinMaxScaler(feature_range=(-1, 1)),
    'NOEMP': RobustScaler(),
    'AAGE': StandardScaler(),
    'CAPGAIN': FunctionTransformer(lambda x: np.log10(x.where(x != 0, 10)), validate=False),
    'CAPLOSS': FunctionTransformer(lambda x: np.log10(x.where(x != 0, 10)), validate=False)
}

preprocessor = ColumnTransformer(
    transformers=[(col, scalers[col], [col]) for col in columns_to_normalize])

pipeline = Pipeline([
    ('preprocessor', preprocessor)
])

df_normalize = pipeline.fit_transform(df_cont)

df_normalize = pd.DataFrame(df_normalize, columns=columns_to_normalize)

df_normalize
```
