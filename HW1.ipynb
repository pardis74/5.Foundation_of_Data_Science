```python
# utility functions
def missing_perc(feat_values):
#     print (feat_values[feat_values.isna()])
    return 100*feat_values.isna().sum() / feat_values.size

def get_cardinality(feat_values):
    return len(np.unique(feat_values.dropna()))


def get_mode_attributes(feat_values):
    value, counts = np.unique(values.dropna(), return_counts=True)
    sortedc = counts.copy()
    sortedc.sort()
    return value[counts == sortedc[-1]][0], sortedc[-1], 100*sortedc[-1] / sortedc.sum()

def get_2mode_attributes(feat_values):
    value, counts = np.unique(values.dropna(), return_counts=True)
    sortedc = counts.copy()
    sortedc.sort()
    return value[counts == sortedc[-2]][0], sortedc[-2], 100*sortedc[-2] / sortedc.sum()
    
def get_mean(feat_values):
    return np.mean(feat_values)

def get_std(feat_values):
    return np.std(feat_values)

def get_min(feat_values):
    return np.min(feat_values)

def get_max(feat_values):
    return np.max(feat_values)

def get_q1(feat_values):
    return feat_values.quantile(0.25)

def get_q3(feat_values):
    return feat_values.quantile(0.75)

def get_median(feat_values):
    return np.median(feat_values)
```

```python
# build categorical DQR dictionary and create a dataframe from it
cat_dqr_val = {}

for feat, values in df_cat.iteritems():
    dqr_val_dict = {"Feature":feat, 
                    'Desc':q2_df.loc[feat]['desc'],
                    'Count':values.dropna().size,
                    '%Missing': missing_perc(values),
                    'Card.':get_cardinality(values),
                   }
    mode, mode_freq, mode_perc = get_mode_attributes(values)
    mode2nd, mode2nd_freq, mode2nd_perc = get_2mode_attributes(values)
    dqr_val_dict['Mode'] = mode
    dqr_val_dict['Mode Freq.'] = mode_freq
    dqr_val_dict['Mode Perc.'] = mode_perc
    dqr_val_dict['2nd Mode'] = mode2nd
    dqr_val_dict['2nd Mode Freq.'] = mode2nd_freq
    dqr_val_dict['2nd Mode Perc.'] = mode2nd_perc
    dqr_val_dict['Note']=''
    
    cat_dqr_val[feat] = dqr_val_dict
    
cat_dqr_df = pd.DataFrame(data=cat_dqr_val).T
cat_dqr_df
```

### Q3 Outlier Identification (25 pts)

#### Q3.1 For each continuous feature, identify the outliers using the IQR method. (15 pts)
For each feature report the lower and upper bounds and number of instances that are identified as outliers.
Then, display boxplots and discuss if your outliers analysis makes sense. Discuss why you would (or would not) use the IQR method.


```python
def get_q1(feat_values):
    return feat_values.quantile(0.25)

def get_q3(feat_values):
    return feat_values.quantile(0.75)

def get_iqr(feat_values):
    return feat_values.quantile(0.75) - feat_values.quantile(0.25)


def get_iqr_upper(feat_values):
    return feat_values.quantile(0.75) + 1.5*(feat_values.quantile(0.75) - feat_values.quantile(0.25))

def get_iqr_lower(feat_values):
    return feat_values.quantile(0.25) - 1.5*(feat_values.quantile(0.75) - feat_values.quantile(0.25))

def get_iqr_outliers(feat_values):
    return (feat_values < get_iqr_lower(feat_values)) | (feat_values > get_iqr_upper(feat_values))


# df_cont.boxplot()
for col in df_cont.columns:
    print(col)
    print( "Upper Bound:", get_iqr_upper (df_cont[col]) )
    print( "Lower Bound:", get_iqr_lower (df_cont[col]) )
    print( "number of outliers:", len(df_cont.loc[get_iqr_outliers(df_cont[col]), [col]]) )
    plt.figure()   # plots figure for each iteration
    df_cont[[col]].boxplot()
    

# df.loc[get_iqr_outliers(df_cont['AAGE']), ['AAGE']]

# df_cont['AAGE'] < get_iqr_lower(df_cont['AAGE'])

# df_cont.loc[]
```
